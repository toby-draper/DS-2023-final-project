# Description
This infographic explores the variation of musical keys across different music genres, and the potential reasons for the distributions found. In conducting this analysis, I aimed to uncover patterns and trends in key usage across various genres, searching for trends in how musical keys contribute to the overall feel and style of music. The infographic presents a series of visualizations that highlight the most common keys used in each genre, as well potential explanations for the trends observed. The data used in this project is a kaggle data set containing the author's listening history from 2022. It consists of just over 10,000 unique tracks and 22 features.
